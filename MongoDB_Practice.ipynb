```python
import pymongo
from pprint import pprint
```

```python
conn = "mongodb://localhost:27017"
client = pymongo.MongoClient(conn)

db = client.yelp_db
collection = db.business
```

```python
businesses = collection.find_one({'city':'Phoenix'})
pprint(businesses)
```

```python
cities = collection.distinct('city')
pprint(cities)
```

```python
states = collection.distinct('state')
print(states)
```

```python
four_stars = collection.find( {'state': 'CA', 'stars': { "$gt": 4} }).limit(5)
for item in four_stars:
    pprint(f"{item['name']} has a rating of {item['stars']}")
```
