```python
# Import dependencies
import requests
import numpy as np
import pandas as pd
import json
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline
import collections
import os
```

```python
business_file = "yelp_academic_dataset_business.json"
```

```python
df = pd.read_json(business_file, lines=True)
```

```python
df.head()
```

```python
reduced_df = df.loc[:, ["name", "business_id", "city","state", "review_count", "stars", "is_open"]]
reduced_df.head()
```

```python
reduced_df.count()
```

```python
open_df = reduced_df.loc[(reduced_df["is_open"] == 1)]
open_df.head()
```

```python
open_df.count()
```

```python
open_df['city'].value_counts()
```

```python
top_10_US = ["Las Vegas", "Phoenix", "Charlotte", "Scottsdale", "Pittsburgh", 
             "Mesa", "Henderson", "Tempe", "Chandler", "Gilbert"]
```

```python
criteria = open_df['city'].isin(["Las Vegas", "Phoenix", "Charlotte", "Scottsdale", "Pittsburgh", 
             "Mesa", "Henderson", "Tempe", "Chandler", "Gilbert"])

top_10_df = open_df[criteria]

top_10_df = top_10_df.reset_index()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
top_10_df.to_csv("alldata.csv", index=False, header=True)
top_10_df.head()
```

```python
business_name_df = top_10_df.loc[:, ["name","city","state"]]
business_name_df.head()
```

```python
business_name_df.to_csv("business_name.csv", index=False, header=True)
business_name_df.head()
```
